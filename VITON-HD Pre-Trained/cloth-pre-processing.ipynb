```python
import os
import io
from PIL import Image
from rembg import remove

def generate_cloth_mask(input_path, output_path):
    """
    Takes an input image of a clothing item and generates a black-and-white
    mask by removing the background.

    Args:
        input_path (str): The path to the source cloth image.
        output_path (str): The path where the generated mask will be saved.
    """
    # --- 1. Validate Input ---
    if not os.path.exists(input_path):
        print(f"❌ Error: Input file not found at '{input_path}'")
        return

    print(f"Processing '{os.path.basename(input_path)}'...")

    try:
        # --- 2. Read the input image file ---
        with open(input_path, 'rb') as f:
            input_bytes = f.read()

        # --- 3. Use rembg to remove the background ---
        # The result is an RGBA image in bytes, where the background is transparent.
        output_bytes = remove(input_bytes)

        # --- 4. Convert bytes to a PIL Image and extract the mask ---
        # Open the RGBA image from the bytes
        rgba_image = Image.open(io.BytesIO(output_bytes)).convert("RGBA")
        
        # The alpha channel ('A') of the RGBA image is our mask.
        # In the alpha channel, the foreground (cloth) is white and the background is black.
        mask = rgba_image.split()[3]

        # --- 5. Save the final mask ---
        # Ensure the output directory exists
        output_dir = os.path.dirname(output_path)
        if output_dir:
            os.makedirs(output_dir, exist_ok=True)
            
        mask.save(output_path)
        print(f"✅ Successfully saved mask to '{output_path}'")

    except Exception as e:
        print(f"❌ An error occurred: {e}")


# --- Main execution part ---

# 1. DEFINE YOUR INPUT AND OUTPUT PATHS HERE
input_image_path = "/kaggle/input/clothes-tryon-dataset/clothes_tryon_dataset/train/cloth/00015_00.jpg"  # <-- This matches your screenshot
output_mask_path = "/kaggle/working/cloth_mask.png" # <-- CORRECTED: This is now a full filename

# 2. RUN THE FUNCTION
# The script will now use the paths you defined above.
print("--- Starting Cloth Mask Generation ---")
generate_cloth_mask(input_image_path, output_mask_path)
print("--- Script Finished ---")
```

```python
import os
import torch
import numpy as np
from PIL import Image
import torchvision.transforms as transforms

# Import the necessary components from the 'schp' library
from schp.modeling.model import SCHP
# 'transform_logits' was removed as it was unused.

def get_palette(num_cls):
    """ Returns the color palette for visualizing the segmentation map"""
    n = num_cls
    palette = [0] * (n * 3)
    for j in range(0, n):
        lab = j
        palette[j * 3 + 0] = 0
        palette[j * 3 + 1] = 0
        palette[j * 3 + 2] = 0
        i = 0
        while lab > 0:
            palette[j * 3 + 0] |= (((lab >> 0) & 1) << (7 - i))
            palette[j * 3 + 1] |= (((lab >> 1) & 1) << (7 - i))
            palette[j * 3 + 2] |= (((lab >> 2) & 1) << (7 - i))
            i += 1
            lab >>= 3
    return palette

def generate_segmentation_map(input_image_path, model_path, output_path):
    """
    Generates a color-coded human segmentation map for a given person image
    using the SCHP model.

    Args:
        input_image_path (str): Path to the input person image.
        model_path (str): Path to the pre-trained SCHP model weights.
        output_path (str): Path to save the generated segmentation map.
    """
    # --- 1. Setup model and transformations ---
    # CORRECTION: Automatically detect and use GPU if available, else CPU.
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # Define the image transformations
    input_transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
    ])
    
    # Load the pre-trained model
    model = SCHP(num_classes=20)
    # CORRECTION: Load model directly to the determined device.
    model.load_state_dict(torch.load(model_path, map_location=device)['state_dict'])
    # CORRECTION: Move model to the determined device.
    model.to(device)
    model.eval()

    # --- 2. Process the image ---
    print(f"Processing '{os.path.basename(input_image_path)}'...")
    try:
        img = Image.open(input_image_path).convert("RGB")
        img_size = img.size
        # CORRECTION: Move the input tensor to the determined device.
        img_tensor = input_transform(img).unsqueeze(0).to(device)

        # Run inference
        with torch.no_grad():
            outputs = model(img_tensor)
        
        # Upsample the output to the original image size
        # The output from the model is a list of tensors, we take the last one.
        upsample_output = torch.nn.functional.interpolate(
            outputs[0][-1], size=img_size, mode='bilinear', align_corners=True
        )
        
        # Get the final parsing result
        parsing_result = torch.argmax(upsample_output, 1).squeeze(0).cpu().numpy().astype(np.uint8)

        # --- 3. Save the result as a color-coded PNG ---
        # CORRECTION: Ensure the output directory exists before saving.
        output_dir = os.path.dirname(output_path)
        if output_dir:
            os.makedirs(output_dir, exist_ok=True)
            
        output_img = Image.fromarray(parsing_result, mode='P') # 'P' mode for palette
        palette = get_palette(20) # 20 classes for the LIP dataset
        output_img.putpalette(palette)
        output_img.save(output_path)
        
        print(f"✅ Successfully saved segmentation map to '{output_path}'")

    except FileNotFoundError:
        print(f"❌ Error: Input file not found at '{input_image_path}'")
    except Exception as e:
        print(f"❌ An error occurred: {e}")


# --- Main execution part ---

# 1. DEFINE YOUR INPUT, MODEL, AND OUTPUT PATHS HERE
input_person_path = "/kaggle/input/model-cloth/model.png" # <-- CHANGE THIS
parsing_model_path = "/kaggle/working/body_pose_model.pth" # <-- CHANGE THIS
# CORRECTION: Changed to a safer default path. This will save the image in your current working directory.
output_parse_path = "./image-parse-v3.png" # <-- CHANGE THIS

# 2. RUN THE FUNCTION
print("--- Starting Human Segmentation ---")
generate_segmentation_map(input_person_path, parsing_model_path, output_parse_path)
print("--- Script Finished ---")
```
